```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score #deklarasi
```

PEMANGGILAN DATASET

```python
marcel = pd.read_csv('smartphone.csv', ';')
marcel
```

```python
marcel = marcel[['Bulan April 2020', 'Bulan Mei 2020', 'Bulan Juni 2020']]
data = np.array(marcel)
print(data)
```

Menampilkan data titik atau Scater

```python
plt.scatter(marcel["Bulan April 2020"], marcel["Bulan Mei 2020"], c='orange', label='Data Points')
plt.xlabel("Bulan April 2020")
plt.ylabel("Bulan Mei 2020")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

```python
x_array = np.array(marcel)
```

menentukan dan mengkonfigurasi fngsi k-means  || menentukan nilai k dari data Memasukkan data trainning pada fungsi klasifikasi

```python
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the KMeans model before accessing cluster_centers_
kmeans.fit(x_array)

centers = kmeans.cluster_centers_
output = plt.scatter(x_array[:, 0], x_array[:, 1], s=50, c=kmeans.labels_)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=50)
plt.title("KMEANS Clustering")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(output)
plt.show()
```

```python
fm = davies_bouldin_score(x_array, kmeans.labels_)
print("Davies-Bouldin Score:", fm)
```
